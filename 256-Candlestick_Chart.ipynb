```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file   

start=datetime.datetime(2021,3,12)
end=datetime.datetime(2021,4,1)

df=data.DataReader(name="GOOG",data_source="yahoo",start=start,end=end)
df
```

```python
###### Here define a function to devide the increase parts and decrease into two parts
###### function increase $ decrease = inc_dec
def inc_dec(c, o):
    if c > o:
        value="Increase"
    elif c < o:
        value="Decrease"
    else:
        value="Equal"
    return value

df["Status"]=[inc_dec(c,o) for c, o in zip(df.Close, df.Open)]
```

```python
df["Middle"]=(df.Open+df.Close)/2
df["Height"]=abs(df.Close-df.Open)               #abs： makes sure that you will always get positive value 
```

```python
df
```

```python
p=figure(x_axis_type='datetime', width=1000, height=300)            #store the figure object in a p variable figure 
p.title="Candlestick Chart"

hours_12=12*60*60*1000
     
p.rect(df.index[df.Status=="Increase"], df.Middle[df.Status=="Increase"],                                        #定义单个方块图的左边位置，纵轴的中点是起始与结束时间的二分之一， 横轴位置以千分之一秒的刻度
       hours_12, df.Height[df.Status=="Increase"], fill_color="green", line_color="black")

p.rect(df.index[df.Status=="Decrease"], df.Middle[df.Status=="Decrease"],  
       hours_12, df.Height[df.Status=="Decrease"], fill_color="red", line_color="black")

output_file("CS.html")
show(p)
                
    
```
